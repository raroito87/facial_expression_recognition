```python
from utils import Preprocessing
import pandas as pd
```

```python
prep = Preprocessing('fer2013')
```

```python
prep.load_data('fer2013.csv', name='alldata').head()
```

```python
prep.get(name='alldata')['Usage'].unique().tolist()
```

```python
pixels=prep.get(name='alldata')['pixels']
```

```python
pixels = pixels.str.split(' ', n=48*48, expand = True)
```

```python
pixels.head()
```

```python
pixels_norm = pixels.applymap(lambda x: 2*(float(x)/255 - 0.5))
pixels_norm.tail()
```

```python
data_norm = pixels_norm.join(prep.load_data('fer2013.csv', name='alldata')['emotion']).join(prep.load_data('fer2013.csv', name='alldata')['Usage'])
data_norm.head()
```

```python
usage_values = data_norm['Usage'].unique().tolist()
usage_values
```

```python
train = data_norm.loc[data_norm['Usage'] == 'Training']
```

```python
train.tail()
```

```python
prep.set(name='train_norm_centered', value = train.drop(columns = ['Usage']))
prep.save(name='train_norm_centered')
```

```python
publicTest = data_norm.loc[data_norm['Usage'] == 'PublicTest']
```

```python
prep.set(name='test_public_norm_centered', value = publicTest.drop(columns = ['Usage']))
prep.save(name='test_public_norm_centered')
```

```python
privateTest = data_norm.loc[data_norm['Usage'] == 'PrivateTest']
```

```python
prep.set(name='test_private_norm_centered', value = privateTest.drop(columns = ['Usage']))
prep.save(name='test_private_norm_centered')
```
