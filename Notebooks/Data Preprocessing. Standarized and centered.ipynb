```python
from utils import Preprocessing
import pandas as pd
import numpy as np
```

```python
prep = Preprocessing('fer2013')
```

```python
prep.load_data('fer2013.csv', name='alldata').head()
```

```python
prep.get(name='alldata')['Usage'].unique().tolist()
```

```python
pixels=prep.get(name='alldata')['pixels']
```

```python
pixels = pixels.str.split(' ', n=48*48, expand = True)
```

```python
pixels.head()
```

```python
x = pixels.values.astype(np.float)
x
```

```python
x -= np.mean(x, axis=0)
x /= np.std(x, axis=0)
```

```python
x
```

```python
pixels_std = pd.DataFrame(x)
pixels_std.tail()
```

```python
data_norm = pixels_std.join(prep.load_data('fer2013.csv', name='alldata')['emotion']).join(prep.load_data('fer2013.csv', name='alldata')['Usage'])
data_norm.head()
```

```python
usage_values = data_norm['Usage'].unique().tolist()
usage_values
```

```python
train = data_norm.loc[data_norm['Usage'] == 'Training']
```

```python
train.tail()
```

```python
prep.set(name='train_std', value = train.drop(columns = ['Usage']))
prep.save(name='train_std')
```

```python
publicTest = data_norm.loc[data_norm['Usage'] == 'PublicTest']
```

```python
prep.set(name='test_public_std', value = publicTest.drop(columns = ['Usage']))
prep.save(name='test_public_std')
```

```python
privateTest = data_norm.loc[data_norm['Usage'] == 'PrivateTest']
```

```python
prep.set(name='test_private_std', value = privateTest.drop(columns = ['Usage']))
prep.save(name='test_private_std')
```
