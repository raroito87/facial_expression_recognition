```python
from preprocessing_utils import Preprocessing
```

```python
prep = Preprocessing('FacialExpression')
```

```python
prep.load_data('fer2013.csv', name='alldata')
```

```python
prep.load_data('fer2013.csv', name='alldata')['Usage'].unique().tolist()
```

```python
pixels=prep.get(name='alldata')['pixels']
```

```python
pixels = pixels.str.split(' ', n=48*48, expand = True)
```

```python
pixels
```

```python
data = pixels.join(prep.load_data('fer2013.csv', name='alldata')['emotion']).join(prep.load_data('fer2013.csv', name='alldata')['Usage'])
```

```python
data.head()
```

```python
prep.set(name='cleaned', value=data)
prep.save(name='cleaned')
```

```python
usage_values = prep.get(name='cleaned')['Usage'].unique().tolist()
usage_values
```

```python
train = prep.get(name='cleaned').loc[prep.get(name='cleaned')['Usage'] == 'Training']
```

```python
train.tail()
```

```python
prep.set(name='train', value = train.drop(columns = ['Usage']))
prep.save(name='train')
```
